```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df_reset=df.reset_index(drop=True)
df_reset.index=df_reset.index+1
df_reset.head()
```

```python
sns.countplot(x='win_by_runs',data=df_reset)
```

```python
df['city'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
sns.distplot(x=df['win_by_wickets'])
```

```python
p=pd.crosstab(df_reset['winner'], df_reset['win_by_runs'])
print(p)
```

```python
sns.histplot(x='city', hue='win_by_runs', data=df_reset)
```

```python
numeric_data = df_reset.select_dtypes(include=["int64", "float64"])
corr_matrix = numeric_data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Heatmap: Correlation Matrix")
plt.show()
```
